```python
base_TTN_path = "Offline-online/Best_base_TTN_results/Training_fqi_offline_online_env_Mountaincar_mem_size_50000_hyper_num_90_date_12_01_2022/_initialbatch_hyperparam_final_avgepisodes.npy"
TTN_Qlearnin_batch_path = "Offline-online/DQN_update_on_batch_32/Training_fqi_offline_online_env_Mountaincar_mem_size_50000_hyper_num_0_date_26_01_2022/_initialbatch_hyperparam_final_avgepisodes.npy"
TTN_Qlearnin_complete_path = "Offline-online/DQN_update_on_complete_data/Training_fqi_offline_online_env_Mountaincar_mem_size_50000_hyper_num_0_date_26_01_2022/_initialbatch_hyperparam_final_avgepisodes.npy"
# TTN_Qlearnin_complete_path = "Offline-online/Training_fqi_offline_online_env_Mountaincar_mem_size_50000_hyper_num_90_date_20_01_2022/_initialbatch_hyperparam_final_avgepisodes.npy"
```

```python
base_TTN = np.load(base_TTN_path)
TTN_Qlearnin_batch = np.load(TTN_Qlearnin_batch_path)
TTN_Qlearnin_complete = np.load(TTN_Qlearnin_complete_path)
# random_sparse_input_128 = np.load(random_sparse_input_path_128)
```

```python
print(f"base_TTN: {base_TTN.shape}")
print(f"TTN_Qlearnin_batch: {TTN_Qlearnin_batch.shape}")
print(f"TTN_Qlearnin_complete: {TTN_Qlearnin_complete.shape}")
# print(f"random_sparse_input_128: {random_sparse_input_128.shape}")
```

```python
base_TTN_mean = np.mean(base_TTN)
print(f"base_TTN_mean: {base_TTN_mean}")

TTN_Qlearnin_batch_mean = np.mean(TTN_Qlearnin_batch)
print(f"TTN_Qlearnin_batch_mean: {TTN_Qlearnin_batch_mean}")

TTN_Qlearnin_complete_mean = np.mean(TTN_Qlearnin_complete)
print(f"TTN_Qlearnin_complete_mean: {TTN_Qlearnin_complete_mean}")

# random_sparse_input_128_mean = np.mean(random_sparse_input_128)
# print(f"random_sparse_input_mean: {random_sparse_input_128_mean}")
```

```python
base_TTN_median = np.median(base_TTN)
print(f"base_TTN_median: {base_TTN_median}")

TTN_Qlearnin_batch_median = np.median(TTN_Qlearnin_batch)
print(f"TTN_Qlearnin_batch_median: {TTN_Qlearnin_batch_median}")

TTN_Qlearnin_complete_median = np.median(TTN_Qlearnin_complete)
print(f"TTN_Qlearnin_complete_median: {TTN_Qlearnin_complete_median}")

# random_sparse_input_128_median = np.median(random_sparse_input_128)
# print(f"random_sparse_input_128_median: {random_sparse_input_128_median}")
```

```python
# plot best results

plt.rcParams['text.usetex'] = True
plt.rcParams["figure.figsize"] = (20,10)

w = 10
base_TTN_mv = np.array([np.mean(base_TTN[i:i+w]) for i in range(0, len(base_TTN))])
TTN_Qlearnin_batch_mv = np.array([np.mean(TTN_Qlearnin_batch[i:i+w]) for i in range(0, len(TTN_Qlearnin_batch))])
TTN_Qlearnin_complete_mv = np.array([np.mean(TTN_Qlearnin_complete[i:i+w]) for i in range(0, len(TTN_Qlearnin_complete))])
# random_sparse_input_128_mv = np.array([np.mean(random_sparse_input_128[i:i+w]) for i in range(0, len(random_sparse_input_128))])



# plots
plt.plot(base_TTN_mv, color='tab:gray', label='Best-TTN')
plt.plot(TTN_Qlearnin_batch_mv, color='tab:red', label='TTN-Rep-Qlearn-Policy-batch-update')
plt.plot(TTN_Qlearnin_complete_mv, color='tab:orange', label='TTN-Rep-Qlearn-Policy-complete-update')
# plt.plot(random_sparse_input_128_mv, color='tab:orange', label='Random-sparse-input-128')



plt.legend()
plt.title(f"TTN vs TTN-Qlearn-batch vs TTN-Qlearn-complete")
plt.ylabel(f"Avg. Steps/Episodes")
plt.xlabel("No. of Episodes")
sns.set(rc={'text.usetex': True})
plt.savefig("TTN vs TTN-Qlearn-batch vs TTN-Qlearn-complete.png", dpi=300)
plt.show()
```
